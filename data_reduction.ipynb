```python

#todo reduce colm from lpg dataset

lpg_price_india = lpg_price_india[['date','delhi_price','mumbai_price','kolkata_price','chennai_price','bengaluru_price']]
```

```python

#^ add average price colm
cities = ['delhi_price','mumbai_price','kolkata_price','chennai_price','bengaluru_price']

lpg_price_india['avg_price'] = lpg_price_india[cities].mean(axis=1)
```

```python
fuel_price_india = fuel_price_india[['date', 'delhi_petrol', 'delhi_diesel', 'mumbai_petrol','mumbai_diesel', 'kolkata_petrol', 'kolkata_diesel', 'chennai_petrol', 'chennai_diesel', 'bengaluru_petrol', 'bengaluru_diesel']]
```

```python

#? add avg price colm for petrol and diesel

petrol_cities = ['delhi_petrol', 'mumbai_petrol', 'kolkata_petrol', 'chennai_petrol', 'bengaluru_petrol']
diesel_cities = ['delhi_diesel', 'mumbai_diesel', 'kolkata_diesel', 'chennai_diesel', 'bengaluru_diesel']

fuel_price_india['avg_petrol'] = fuel_price_india[petrol_cities].mean(axis=1).round(2)
fuel_price_india['avg_diesel'] = fuel_price_india[diesel_cities].mean(axis=1).round(2)
```
