3. Data Integration

```python
import pandas as pd
```

```python

#? load datasets
brent_oil = pd.read_csv('C:/Users/anasq\OneDrive/Desktop/PetroAnalyst/Dataset/brent_oil_price.csv')
wti_oil = pd.read_csv('C:/Users/anasq/OneDrive/Desktop/PetroAnalyst/Dataset/wti_crude_oil_price.csv')
lpg_price_india = pd.read_csv('C:/Users/anasq/OneDrive/Desktop/PetroAnalyst/Dataset/reduced_lpg_price.csv')
fuel_price_india = pd.read_csv('C:/Users/anasq/OneDrive/Desktop/PetroAnalyst/Dataset/reduced_fuel_price.csv')

GPR_index = pd.read_csv('C:/Users/anasq/OneDrive/Desktop/PetroAnalyst/Dataset/reduced_gpr.csv')
INR_to_USD = pd.read_csv('C:/Users/anasq/OneDrive/Desktop/PetroAnalyst/Dataset/INR_to_USD.csv')
```

```python

#todo Convert each date column to datetime (respective formats)
brent_oil['date'] = pd.to_datetime(brent_oil['date'])                        
wti_oil['date']   = pd.to_datetime(wti_oil['date'])                          
INR_to_USD['date']   = pd.to_datetime(INR_to_USD['date'])                     
GPR_index['date'] = pd.to_datetime(GPR_index['date'])        
lpg_price_india['date']  = pd.to_datetime(lpg_price_india['date'], dayfirst=True)     
fuel_price_india['date'] = pd.to_datetime(fuel_price_india['date'], dayfirst=True)     

#^ Normalize ALL to YYYY/MM/DD string format
for df in [brent_oil, wti_oil, INR_to_USD, GPR_index, lpg_price_india, fuel_price_india]:
    df['date'] = df['date'].dt.strftime('%Y/%m/%d')
```

```python

#~ just verification

print(brent_oil['date'].head(3), '\n')
print(wti_oil['date'].head(3), '\n')
print(INR_to_USD['date'].head(3), '\n')
print(GPR_index['date'].head(3), '\n')
print(lpg_price_india['date'].head(3), '\n')
print(fuel_price_india['date'].head(3), '\n')
```

```python

#& Merge all 6 datasets on date 
master = brent_oil.merge(wti_oil,  on='date') \
              .merge(INR_to_USD,  on='date') \
              .merge(GPR_index,  on='date') \
              .merge(lpg_price_india,  on='date') \
              .merge(fuel_price_india, on='date')

#! remove unknown columns
master = master.loc[:, ~master.columns.str.contains('^Unnamed')]

#~ Verify
print("Shape:", master.shape)        
print("Columns:", list(master.columns))
print("Nulls:", master.isnull().sum())
print(master.head(3))
```
