```python

#~ list of all datasets
datasets = [brent_oil, wti_oil, lpg_price_india, fuel_price_india, GPR_index, INR_to_USD]

for df in datasets:
    print(df.columns)
    print('\n')
```

```python

#todo statistical analysis
for df in datasets:
    print(df.describe())
    print('\n')
```

```python

#^ Check data types and basic info
for i, df in enumerate(datasets):
    print(f"\n{'='*50}")
    print(f"Dataset {i+1} Info:")
    print(f"{'='*50}")
    df.info()
    print(f"Shape: {df.shape}")
```

```python

#todo Check for missing values 
print("\n" + "="*50)
print("MISSING VALUES")
print("="*50)
for i, df in enumerate(datasets):
    missing = df.isnull().sum()
    if missing.sum() > 0:
        print(f"\nDataset {i+1} - Missing values:")
        print(missing[missing > 0])
    else:
        print(f"\nDataset {i+1} - No missing values ✓")
```

```python

#^ Check for duplicates
print("\n" + "="*50)
print("DUPLICATES")
print("="*50)
for i, df in enumerate(datasets):
    dup_count = df.duplicated().sum()
    print(f"Dataset {i+1} - Duplicate rows: {dup_count}")
```
